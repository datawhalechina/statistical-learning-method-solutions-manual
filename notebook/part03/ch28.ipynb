```python
def minmax_function(A):
    """
    ä»æ”¶ç›ŠçŸ©é˜µä¸­è®¡ç®—minmaxçš„ç®—æ³•
    :param A: æ”¶ç›ŠçŸ©é˜µ
    :return: è®¡ç®—å¾—åˆ°çš„minmaxç»“æœ
    """
    index_max = []
    for i in range(len(A)):
        # è®¡ç®—æ¯ä¸€è¡Œçš„æœ€å¤§å€¼
        index_max.append(A[i,:].max())
    
    # è®¡ç®—æ¯ä¸€è¡Œçš„æœ€å¤§å€¼ä¸­çš„æœ€å°å€¼
    minmax = min(index_max)
    return minmax
```

```python
def maxmin_function(A):
    """
    ä»æ”¶ç›ŠçŸ©é˜µä¸­è®¡ç®—maxminçš„ç®—æ³•
    :param A: æ”¶ç›ŠçŸ©é˜µ
    :return: è®¡ç®—å¾—åˆ°çš„maxminç»“æœ
    """
    column_min = []
    for i in range(len(A)):
        # è®¡ç®—æ¯ä¸€åˆ—çš„æœ€å°å€¼
        column_min.append(A[:,i].min())
        
    # è®¡ç®—æ¯ä¸€åˆ—çš„æœ€å°å€¼ä¸­çš„æœ€å¤§å€¼
    maxmin = max(column_min)
    return maxmin
```

```python
# åˆ›å»ºæ”¶ç›ŠçŸ©é˜µ
A = np.array([[-1,2],[4,1]])
# è®¡ç®—maxmin
maxmin = maxmin_function(A)
# è®¡ç®—minmax
minmax = minmax_function(A)
# è¾“å‡ºç»“æœ
print("maxmin =", maxmin)
print("minmax =", minmax)
```
